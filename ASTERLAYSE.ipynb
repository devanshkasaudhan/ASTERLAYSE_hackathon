Certainly! Here's a detailed description of each of the features you've listed:

1. **Object full name**: This is the full name of the asteroid, which may include additional information about its discovery and observation history.

2. **a (semi-major axis)**: The semi-major axis is a fundamental parameter in describing the shape and size of an asteroid's orbit. It represents half of the longest diameter of the elliptical orbit an asteroid follows around the Sun, measured in astronomical units (au).

3. **e (eccentricity)**: Eccentricity describes the shape of an asteroid's orbit. It is a measure of how elongated or circular the orbit is. An eccentricity of 0 indicates a perfectly circular orbit, while values between 0 and 1 represent increasingly elongated, elliptical orbits.

4. **i (Inclination with respect to x-y ecliptic plane)**: Inclination is the angle between an asteroid's orbital plane and the plane of the Earth's orbit around the Sun (the ecliptic plane). It is measured in degrees.

5. **om (Longitude of the ascending node)**: This is the angle measured along the ecliptic plane from the vernal equinox to the point where an asteroid's orbit crosses the ecliptic plane in a northward direction.

6. **w (argument of perihelion)**: The argument of perihelion is the angle measured from the ascending node to the point where an asteroid is closest to the Sun in its orbit, known as perihelion.

7. **q (perihelion distance)**: The perihelion distance represents the closest approach of an asteroid to the Sun in its orbit, measured in astronomical units (au).

8. **ad (aphelion distance)**: The aphelion distance is the farthest point from the Sun in an asteroid's orbit, measured in astronomical units (au).

9. **per_y (Orbital period in years)**: This is the time it takes for the asteroid to complete one orbit around the Sun, measured in years.

10. **data_arc (data arc-span)**: Data arc-span is the duration, in days, over which observations and data for the asteroid's orbit have been collected.

11. **condition_code (Orbit condition code)**: The orbit condition code is a numerical indicator of the accuracy and reliability of an asteroid's orbital parameters. Lower values generally indicate more reliable orbits.

12. **n_obs_used (Number of observations used)**: This represents the total number of observations that have been used to calculate and refine the asteroid's orbital parameters.

13. **H (Absolute Magnitude parameter)**: The absolute magnitude is a measure of an asteroid's intrinsic brightness, corrected for its distance from both the Sun and Earth.

14. **neo (Near Earth Object)**: A binary indicator, typically "Y" or "N," that denotes whether the asteroid is classified as a Near Earth Object, which means it has an orbit that brings it into proximity with Earth's orbit.

15. **pha (Physically Hazardous Asteroid)**: Similar to "neo," this is a binary indicator ("Y" or "N") that denotes whether the asteroid is classified as a potentially hazardous asteroid based on its size and orbit.

16. **diameter**: This is the estimated diameter of the asteroid in kilometers, providing information about its size.

17. **extent**: The object bi/tri axial ellipsoid dimensions represent the dimensions of the asteroid if it were to be approximated as an ellipsoid. It provides information about its shape and size.

18. **albedo (geometric albedo)**: Geometric albedo is a measure of an asteroid's reflectivity, indicating how much sunlight it reflects. It is a unitless value between 0 and 1, with 1 indicating perfect reflectivity.

19. **rot_per (Rotation Period)**: This is the rotation period of the asteroid, which is the time it takes for the asteroid to complete one full rotation on its axis, measured in hours.

20. **GM (Standard gravitational parameter)**: The standard gravitational parameter is a product of the asteroid's mass and the gravitational constant. It is used in calculations related to the asteroid's gravitational effects on other objects.

21. **BV (Color index B-V)**: The B-V color index represents the magnitude difference between an asteroid's brightness in blue (B) and visible (V) light. It provides information about the asteroid's color.

22. **UB (Color index U-B)**: Similar to B-V, the U-B color index represents the magnitude difference between an asteroid's brightness in ultraviolet (U) and blue (B) light.

23. **IR (Color index I-R)**: The I-R color index represents the magnitude difference between an asteroid's brightness in the infrared (I) and red (R) wavelengths.

24. **spec_B (Spectral taxonomic type SMASSII)**: SMASSII is a spectral taxonomic classification system for asteroids based on their reflectance spectra. It categorizes asteroids into different types, such as C, S, or M, based on their spectral features.

25. **spec_T (Spectral taxonomic type Tholen)**: Tholen is another spectral taxonomic classification system for asteroids, providing information about their composition and characteristics.

26. **G (Magnitude slope parameter)**: The magnitude slope parameter is used to model how an asteroid's brightness changes with phase angle.

27. **class (asteroid orbit class)**: This represents the classification of the asteroid's orbit, which can include various categories such as Atira, Amor, Apollo, or Centaur, depending on its orbital characteristics.

28. **n (Mean motion)**: Mean motion represents the angular speed at which an asteroid moves along its orbit, measured in degrees per day.

29. **per (Orbital Period in days)**: This is the time it takes for the asteroid to complete one orbit around the Sun, measured in days.

30. **ma (Mean anomaly)**: Mean anomaly is an angular parameter that describes the asteroid's position in its orbit at a specific time.

31. **moid (Earth Minimum orbit Intersection Distance)**: MOID represents the closest distance between an asteroid's orbit and the Earth's orbit, measured in astronomical units (au). It indicates how closely the asteroid's path intersects with Earth's orbit.

These features are crucial for understanding and characterizing asteroids, their orbits, physical properties, and potential risks to Earth.

## Feature transformation

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
df['neo']= le.fit_transform(df['neo'])
df['pha']= le.fit_transform(df['pha'])
df['condition_code']= le.fit_transform(df['condition_code'])
```

## Imputation

```python
import pandas as pd
from sklearn.impute import KNNImputer
# Select the features used for imputation
impute_features = ['data_arc', 'condition_code', 'H', 'pha']
# Create a KNNImputer with k neighbors (you can adjust k as needed)
knn_imputer = KNNImputer(n_neighbors=5)

# Perform imputation on the selected features
df[impute_features] = knn_imputer.fit_transform(df[impute_features])
```

# Modeling

## train test split of training data

```python
from sklearn.model_selection import train_test_split

x = df.drop(['moid'], axis=1)
y = df['moid']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
x_train.head()
```

## linear regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score

# Create and fit the multiple regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred_LR = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred_LR)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_LR)
r2 = r2_score(y_test, y_pred_LR)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```

## Gradient boosting regression

```python
from sklearn.ensemble import GradientBoostingRegressor

gbc = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

gbc.fit(x_train, y_train)
```

```python
y_pred = gbc.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```

## Hist Gradient Boosting regression

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingRegressor

# Create and train the HistGradientBoostingClassifier model
hgbc = HistGradientBoostingRegressor(max_iter=100, learning_rate=0.1, max_depth=3, random_state=42)
hgbc.fit(x_train, y_train)
```

```python
predictions_hgbc = hgbc.predict(x_test)
mse = mean_squared_error(y_test, predictions_hgbc)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predictions_hgbc)
r2 = r2_score(y_test, predictions_hgbc)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```
